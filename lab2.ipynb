## Visualizations

```python
df = df[['Date Rptd','DATE OCC','TIME OCC','AREA','REGION','Crm Cd Desc']]
df.head()
```

```python
df.hist(layout=(3,3), figsize=(10, 5), sharey=False, sharex=False, bins=50) #use multiple charts to show the distribution of all columns
```

The first graph shows the most common time crimes are reported showing significant peaks at the whole hour values (2100,2200,2300) and the 30 minute marks (2130,2230,2330). This is useful as it shows many reports may round or approximate the time of the crime rather than utilize the exact time. The second graph demostrates the areas which experience the most reports. Primarily Area 1 and Area 12.

```python
sum_time_by_area = df.groupby('TIME OCC').count()
sum_time_by_area.plot()
```

The first set of charts displays histographs of the Time Occured and the Area as reported. The Line chart displays a more detailed breakdown of the time occured data.
